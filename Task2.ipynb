```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('test.csv')

# Display the first few rows of the DataFrame
print("First 5 rows of the data:")
print(df.head())

# Filter rows based on a condition
# Example: Filter rows where a column named 'age' is greater than 30
filtered_df = df[df['px_height'] > 500]
print("Filtered data (age > 30):")
print(filtered_df.head())

# Handling missing values
# Fill missing values with the mean of the column
df_filled = df.fillna(df.mean())
print("Data after filling missing values with column mean:")
print(df_filled.head())

# Drop rows with missing values
df_dropped = df.dropna()
print("Data after dropping rows with missing values:")
print(df_dropped.head())

# Calculate summary statistics
print("Summary statistics of the dataset:")
print(df.describe())
```
